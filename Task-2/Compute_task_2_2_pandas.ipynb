```python
df.sort_values('danceability',ascending=False)[['song_title','artist','danceability']].head(10)

#this shows the most danceable songs
#sort_values is a function used to sort the values of a dataframe according to the given parameter and ascending=False means that the dataframe will be sorted in descending order
#the part in the box bracket is used to define the parameter that are to be shown from the dataframe after the sorting is done
#head(10) is used to show the first 10 rows of the sorted dataframe
```

```python
df.sort_values('duration_ms')[['song_title', 'artist', 'duration_ms']].tail(10)

#this shows us the slowest song. To do so we sorted the dataframe wrt the duration of the song in milliseconds.
#also we used an alternate method of showing the longest song by displaying the tail instead of sorting the dataframe in descending order.
```

```python
#Now we want to filter the dataframe over more than one parameter

filter=df[df['duration_ms'] < 400000]   #this will help us cut down the data to songs that have a duration less than 4 minutes

filter.sort_values('energy',ascending=False)[['song_title','artist','energy']].head(10)   #this sorts the songs with duration less than
```

```python
lessthan3=df[df['duration_ms']<300000]

lessthan3.sort_values('tempo',ascending=False)[['song_title','artist']].head(10)

#using the similar logic as above we first reduce the dataframe to songs which are smaller than 3 seconds
#then we sort the filtered dataframe according to the value of tempo of the song. We then display the name of the song and the artist
```

```python
fil=df[(df['artist'] == 'Drake') & (df['energy'] > 0.5)]
fil.sort_values('energy')[['song_title','energy']]
```

```python
fil1=df[['instrumentalness','song_title','artist']]
fil1.sort_values('instrumentalness',ascending=False)[['song_title','artist','instrumentalness']].head(10)
```

```python
df1=df.groupby('artist')['tempo'].mean()    #groupby will group the data of the tempo according to the distinct entries in the artist column
                                            #mean will find the mean tempo for each artist

df2=df1.sort_values(ascending=False).head(10)   #this newly made dataframe is sorted in descending order and only its first 10 rows are taken.
df3=pd.DataFrame({'Artist': df2.index, 'Average Tempo': df2.values})    # this makes the new dataframe of the singers with the top 10 values of average tempo
df3
```
